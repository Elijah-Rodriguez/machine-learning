# Import Dependencies

```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
import time
import numpy as np
```

# Web Scraping with Requests Library

I'll begin by defining a year range to scrape data from. Then I'll use a defined url variable to use in a for loop that utilizes the requests library to grab the html from the page. The time library is used to create a lag between each request, avoiding sending too many requests to the website at once.

```python
years = list(range(1980, 2026))
```

```python
roy_url = "https://www.basketball-reference.com/awards/awards_{}.html"

for year in years:
  url = roy_url.format(year)
  data = requests.get(url)

  with open(".../roy/roy_{}.html".format(year), "w+") as f:
    f.write(data.text)

  print(f"Scraped rookie of the year data for the {year} season.")
  lag = np.random.uniform(low=5, high=60)
  time.sleep(lag)  
```

Here I define a slightly different range list not including the 2024-2025 season as the voting hasn't happened yet, therefore the page doesn't exist. This loop goes through the scraped HTML files and parses them to grab only the table that is relevant, the table of the Rookie of the Year winners.

```python
nyears = list(range(1980, 2025))
```

```python
dfr = []

for nyear in nyears:
  with open (".../roy/roy_{}.html".format(nyear)) as f:
    page = f.read()

  soup = BeautifulSoup(page, 'html.parser')
  soup.find('tr', class_="over_header").decompose()
  roy_table = soup.find(id="roy")
  roy = pd.read_html(str(roy_table))[0]
  roy["Year"] = nyear

  dfr.append(roy)
```

Combine all the individual dataframes into one master dataframe that holds all the Rookie of the Year data.

```python
roy_stats = pd.concat(dfr)
roy_stats.head()
```

```python
roy_stats.to_csv("roy_dup.csv", index=False)
```

# Team Standings

The logic for this is identical to the previous, except we can use the 2024-2025 standings with no conflicts.

```python
team_st_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"

for year in years:
  url = team_st_url.format(year)
  data = requests.get(url)

  with open(".../standings/standings_{}.html".format(year), "w+") as f:
    f.write(data.text)
  
  print(f"Scraped standings data for the {year} season.")
  lag = np.random.uniform(low=5, high=60)
  time.sleep(lag)
```

Same logic as the Rookie of the Year parsing, just with some more specific cases to tailor towards the structure of the standings table.

```python
dft = []

for year in years:
  with open (".../standings/standings_{}.html".format(year)) as f:
    page = f.read()

  soup = BeautifulSoup(page, "html.parser")
  tHeads = soup.find_all('tr', class_="thead")
  for tHead in tHeads:
    tHead.decompose()
  team_table = soup.find(id="all_divs_standings_E")
  team_st = pd.read_html(str(team_table))[0]
  team_st["Year"] = year
  team_st["Team"] = team_st["Eastern Conference"]
  del team_st["Eastern Conference"]
  dft.append(team_st)

  soup = BeautifulSoup(page, "html.parser")
  tHeads = soup.find_all('tr', class_="thead")
  for tHead in tHeads:
    tHead.decompose()
  team_table = soup.find(id="all_divs_standings_W")
  team_st = pd.read_html(str(team_table))[0]
  team_st["Year"] = year
  team_st["Team"] = team_st["Western Conference"]
  del team_st["Western Conference"]
  dft.append(team_st)
```

```python
teams = pd.concat(dft)
teams.head()
```

```python
teams.to_csv("team_standings.csv", index=False)
```

# Import and Clean Selenium-scraped Data

Same parsing logic as before, just with the stats of all rookies for a given year from the selenium-scraped data.

```python
dfs=[]

for year in years:
  with open (".../rookies/rookie_stats_{}.html".format(year)) as f:
    page = f.read()

  soup = BeautifulSoup(page, "html.parser")
  oHeads = soup.find_all('tr', class_="over_header")
  for oHead in oHeads:
    oHead.decompose()
  rookie_table = soup.find(id="rookies")
  rookies = pd.read_html(str(rookie_table))[0]
  rookies["Year"] = year

  dfs.append(rookies)
```

```python
rookie_stats = pd.concat(dfs)
rookie_stats.head(25)
```

```python
rookie_stats.to_csv("rookie_stats.csv", index=False)
```
